# Sanity check on a known bound-entangled example

Before trusting the detection pipeline on random states, we run it on a state
whose answer is known: the one-parameter Horodecki 3⊗3 bound-entangled state
(`example_bound_entangled` in `common.jl`). It is PPT for every parameter value
and entangled for `0 < a < 1`, so it is exactly the kind of state the PPT²
search is looking for — a good calibration target for the witnesses.

```julia
include("common.jl")
```

## The example state

Confirm it is a valid density operator (PSD) and PPT.

```julia
ent = example_bound_entangled(0.5)
(psd = eigmin(Hermitian(ent)) ≥ -1e-10, ppt = is_ppt(ent, 3, 3))
```

## Detect the entanglement with DPS

Because the state is bound entangled, the standard DPS robustness sits right at
the numerical boundary (≈ 0). Switching the noise model to `separable` gives a
clearly positive robustness, certifying entanglement together with a witness `w`.

```julia
r_std, _ = entanglement_robustness(ent, [3, 3], 2; solver=Mosek.Optimizer)
r_sep, w = entanglement_robustness(ent, [3, 3], 2, noise="separable"; solver=Mosek.Optimizer)
(standard = r_std, separable = r_sep)
```

## Rationalise the witness

Round the numerical witness to exact rationals and confirm it still reports a
negative expectation on the state (⟨w, ent⟩ < 0).

```julia
w_rat = rationalize.(w, tol=1e-5)
tr(w_rat * ent)
```

## The anti-diagonal `B` witness

A second, structural witness (the anti-diagonal ±1 operator) also separates this
family; its expectation on the example is negative.

```julia
tr(B(9) * ent)
```
